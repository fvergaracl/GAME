```python
import requests
import pandas as pd
```

```python
API_URL= "https://gamification.dev.apps.deustotech.eu/api/v1/"

def get_all_games():
    response = requests.get(API_URL + "games")
    response_json = response.json()
    games = pd.DataFrame(response_json['items'])
    page = 1
    while response_json['search_options']['total_count'] > len(games):
        response = requests.get(API_URL + "games", params={"page": page + 1})
        response_json = response.json()
        games = games.append(pd.DataFrame(response_json['items']))
        page += 1
    return games

def get_all_tasks(game_id):
    response = requests.get(API_URL + "games/" + game_id + "/tasks")
    response_json = response.json()
    tasks = pd.DataFrame(response_json['items'])
    page = 1
    while response_json['search_options']['total_count'] > len(tasks):
        response = requests.get(API_URL + "games/" + game_id + "/tasks", params={"page": page + 1})
        response_json = response.json()
        tasks = pd.concat([tasks, pd.DataFrame(response_json['items'])], ignore_index=True)
        page += 1

    return tasks


def get_points_details(game_id, task_id):
    response = requests.get(API_URL + "games/" + game_id + "/tasks/" + task_id + "/points/details")
    if response.status_code != 200:
        return None
    return response.json()
```

```python
all_games = get_all_games()
all_games
game_tasks = []
df_games = pd.DataFrame(all_games)


for row in df_games.iterrows():
    if (row[1]['gameId'] == 'a9eec376-2bf6-440d-8b83-3091fbba2a8c'):
        tasks = get_all_tasks(row[1]['gameId'])
        df_tasks = pd.DataFrame(tasks)
        new_row = row[1]
        for row_task in df_tasks.iterrows():
            new_row['task_id'] = row_task[1]['id']
            new_row['externalTaskId'] = row_task[1]['externalTaskId']
            game_tasks.append(new_row)
    game_tasks_df = pd.DataFrame( 
        game_tasks,
        columns=['gameId','externalTaskId'])

game_tasks_points = []
for row in game_tasks_df.iterrows():
    points_details = get_points_details(row[1]['gameId'], row[1]['externalTaskId'])
    if not (points_details == None) and len (points_details) > 0:
        for points  in points_details:
            game_tasks_points.append({
                'gameId': row[1]['gameId'],
                'externalTaskId': row[1]['externalTaskId'],
                'points': points['points'],
                'externalUserId': points['externalUserId']
            })
    
game_tasks_points_df = pd.DataFrame(
    game_tasks_points,
    columns=['gameId','externalTaskId','points', 'externalUserId'])

```
